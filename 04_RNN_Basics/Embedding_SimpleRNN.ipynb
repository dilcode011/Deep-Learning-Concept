```python
import numpy as np

docs= ['gp india',
       'india india',
       'hip hip hurray',
       'jeetega bhai jeetega india jeetega',
       'bharat mata ki jai',
       'kohli kohli',
       'sachin sachin',
       'dhoni dhoni',
       'dil ji ki jai',
       'inquilab zindabad'
]
```

```python
#oov_token will convert every new word with <nothing>
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_docs
```

```python
tokenizer.document_count
```

```python
#text_to_sequences ---> generate sequence
sequences= tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
```

```python
#post means after --- zero will put after this
sequences =pad_sequences(sequences,padding='post')
```

```python
sequences
```

```python
from tensorflow.keras.layers import Embedding
model=Sequential()
model.add(Embedding(18,output_dim=2,input_length=5))
```

```python
model.compile('adam','accuracy')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

# The vocabulary size is len(tokenizer.word_index) + 1 to account for 0-based indexing and actual word indices
vocab_size = len(tokenizer.word_index) + 1

model = Sequential()
# Removed input_length as it is deprecated. Keras can infer the input shape.
model.add(Embedding(vocab_size, output_dim=2))
model.compile('adam', 'accuracy')

pred = model.predict(sequences)
print(pred)
```

IMDB DATASET

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(x_train,y_train),(x_test,y_test)=imdb.load_data()
```

```python
x_train
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
len(x_train[21]) #length of every text is not similar so we will cut the text
```

```python
#trim the text
#This give only first 50 words of every text
x_train=pad_sequences(x_train,padding='post',maxlen=50)
x_test=pad_sequences(x_test,padding='post',maxlen=50)
```

```python
x_train.shape
```

```python
#return sequences --->
model=Sequential()
model.add(Embedding(10000,2))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```
