```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
#split dataset
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
#padding='valid'
model=Sequential()
#32 filters and 3x3
#Padding='valid' ----> No padding applied
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',input_shape=(28,28,1),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
#padding='same'
model=Sequential()
#32 filters and 3x3
#Padding='same' ----> keras automatically apply padding as per need
#no data loss
model.add(Conv2D(32,kernel_size=(3,3),padding='same',input_shape=(28,28,1),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
#STRIDES
model=Sequential()
#32 filters and 3x3
#strides=(2,2) --->(2=horizontal,2=vertical)
#strides means how it moves
#no data loss
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),input_shape=(28,28,1),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu'))

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
#POOLING
model=Sequential()
#32 filters and 3x3
#strides=(2,2) --->(2=horizontal,2=vertical)
#strides means how it moves
#no data loss
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',strides=(2,2),input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',strides=(2,2),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))


model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```
