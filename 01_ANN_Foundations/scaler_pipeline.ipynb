```python
#Scaling ensures that "Salary" (in thousands) doesn't outweigh "Age" (in tens) just because its numbers are bigger.
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Sample data: Age, Salary, and City
data = {'Age': [25, 45, 35], 'Salary': [50000, 90000, 70000], 'City': ['Delhi', 'Mumbai', 'Delhi']}
df = pd.DataFrame(data)
```

```python
# 1. Label Encoding (Categorical to Numerical)
le = LabelEncoder()
df['City'] = le.fit_transform(df['City'])
```

```python
# 2. Standard Scaling (Mean=0, StdDev=1)
# MANDATORY for ANNs to prevent large values from dominating weights
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
print("Original Data:\n", df)
print("\nScaled Data (Ready for ANN):\n", scaled_data)
```
