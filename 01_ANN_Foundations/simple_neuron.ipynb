```python
import numpy as np
```

```python
# 1. Inputs (Features)
x = np.array([2.0, 3.0])
```

```python
# 2. Parameters (Weights and Bias)
# Weights represent the strength of the connection
# Bias allows the neuron to activate even if inputs are zero
w = np.array([0.5, -0.2])
b = 0.1
```

```python
# 3. Forward Pass Logic
# Z = (x1 * w1 + x2 * w2) + b
z = np.dot(x, w) + b
```

```python
# 4. Activation (Sigmoid)
def sigmoid(val):
    return 1 / (1 + np.exp(-val))
```

```python
output = sigmoid(z)

print(f"Input features: {x}")
print(f"Weighted Sum (Z): {z}")
print(f"Neuron Output: {output:.4f}")
```
