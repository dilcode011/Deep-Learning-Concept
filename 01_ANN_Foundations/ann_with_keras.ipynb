```python
#This is the smallest unit of an ANN. It takes inputs, applies weights to show "importance," adds a bias for flexibility, and uses an activation function to squash the result between 0 and 1.
```

```python
#This file shows the "Sequential" nature of ANNs. It uses Backpropagation (via the optimizer) to reduce the Loss (error) after every training cycle.
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def create_model():
    model = Sequential([
        # 1st Hidden Layer (Input_dim must match your number of features)
        Dense(units=6, activation='relu', input_dim=3),
        
        # 2nd Hidden Layer
        Dense(units=6, activation='relu'),
        
        # Output Layer (Sigmoid for binary classification)
        Dense(units=1, activation='sigmoid')
    ])
```

```python
# Adam is the most popular optimizer; it adapts the learning rate
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
print("ANN Model Summary:")
model = create_model()
model.summary()
```
